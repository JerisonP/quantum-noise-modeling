### Full Derivation of the Rotating-Frame Hamiltonian and Time Evolution of the Density Matrix

Here, we combine explicit steps from both the Hamiltonian transformation and density matrix evolution. This derivation includes the mathematical steps for obtaining the rotating-frame Hamiltonian $H_{\text{R.F.}}$ and the detailed time evolution of the density matrix in the rotating frame.

---

### Derivation of the Rotating-Frame Hamiltonian and Time Evolution of the Density Matrix

We start with the time evolution of the density matrix $\rho$ under the lab-frame Hamiltonian $H_{\text{lab}}$, described by the Lindblad equation:
$$
\frac{d\rho}{dt} = -i[H_{\text{lab}}, \rho].
$$

In the lab frame, the Hamiltonian $H_{\text{lab}}$ is defined as:
$$
H_{\text{lab}} = \frac{1}{2} (\omega_q + \zeta(t)) \sigma_z + f(t) \cos(\omega_d t) \sigma_x,
$$
where:
- $\omega_q$: qubit frequency,
- $\zeta(t)$: time-dependent noise term,
- $f(t)$: drive amplitude,
- $\omega_d$: drive frequency.

To simplify the analysis, we transform to a rotating frame oscillating at the drive frequency $\omega_d$ by applying a unitary transformation $S(t) = \exp(-i \omega_d t \sigma_z / 2)$. The rotating-frame Hamiltonian $H_{\text{R.F.}}$ is given by:
$$
H_{\text{R.F.}} = S^{\dagger}(t) H_{\text{lab}} S(t) - i S^{\dagger}(t) \frac{dS(t)}{dt}.
$$

---

### Step 1: Transforming $H_{\text{lab}}$ to the Rotating Frame

#### 1.1 Transformation of the $\sigma_z$ Term
The first term in $H_{\text{lab}}$, $\frac{1}{2} (\omega_q + \zeta(t)) \sigma_z$, commutes with $S(t)$, so it remains unchanged under the transformation:
$$
S^{\dagger} \left( \frac{1}{2} (\omega_q + \zeta(t)) \sigma_z \right) S = \frac{1}{2} (\omega_q + \zeta(t)) \sigma_z.
$$

#### 1.2 Transformation of the $\sigma_x$ Term
The second term in $H_{\text{lab}}$, $f(t) \cos(\omega_d t) \sigma_x$, does not commute with $S(t)$. Using the identity:
$$
S^{\dagger}(t) \sigma_x S(t) = \cos(\omega_d t) \sigma_x + \sin(\omega_d t) \sigma_y,
$$
we transform the $\sigma_x$ term in $H_{\text{lab}}$:
$$
S^{\dagger}(t) \left( f(t) \cos(\omega_d t) \sigma_x \right) S(t) = f(t) \cos(\omega_d t) \left( \cos(\omega_d t) \sigma_x + \sin(\omega_d t) \sigma_y \right).
$$
Expanding this expression using trigonometric identities gives:
$$
f(t) \cos(\omega_d t) \sigma_x = f(t) \left( \frac{1 + \cos(2 \omega_d t)}{2} \sigma_x + \frac{\sin(2 \omega_d t)}{2} \sigma_y \right).
$$

#### 1.3 Non-Inertial Term
To complete the transformation, we compute the non-inertial term:
$$
-i S^{\dagger}(t) \frac{dS(t)}{dt}.
$$
Calculating $\frac{dS(t)}{dt}$ for $S(t) = \exp(-i \omega_d t \sigma_z / 2)$:
$$
\frac{dS(t)}{dt} = -i \frac{\omega_d}{2} \sigma_z S(t),
$$
so
$$
-i S^{\dagger}(t) \frac{dS(t)}{dt} = \frac{\omega_d}{2} \sigma_z.
$$

---

### Step 2: Combining Terms for $H_{\text{R.F.}}$

The rotating-frame Hamiltonian, combining all terms, is:
$$
H_{\text{R.F.}} = \frac{1}{2} (\omega_q - \omega_d + \zeta(t)) \sigma_z + f(t) \left( \frac{1 + \cos(2 \omega_d t)}{2} \sigma_x + \frac{\sin(2 \omega_d t)}{2} \sigma_y \right).
$$
Applying the Rotating Wave Approximation (RWA), which neglects terms oscillating at $2 \omega_d$, we obtain:
$$
H_{\text{R.F.}} = \frac{1}{2} (\Delta + \zeta(t)) \sigma_z + \frac{f(t)}{2} \sigma_x,
$$
where $\Delta = \omega_q - \omega_d$ is the detuning.

---

### Step 3: Time Evolution of $\rho$ in the Rotating Frame

The time evolution of $\rho$ in the rotating frame is given by:
$$
\frac{d\rho}{dt} = -i [H_{\text{R.F.}}, \rho].
$$

Substituting $H_{\text{R.F.}}$ and $\rho$ explicitly, we write:
$$
H_{\text{R.F.}} = \begin{pmatrix} \frac{1}{2} (\Delta + \zeta(t)) & \frac{f(t)}{2} \\ \frac{f(t)}{2} & -\frac{1}{2} (\Delta + \zeta(t)) \end{pmatrix},
\quad \rho = \begin{pmatrix} \rho_{00} & \rho_{01} \\ \rho_{10} & \rho_{11} \end{pmatrix}.
$$

#### Explicit Calculation of $\frac{d\rho}{dt}$:
The commutator $[H_{\text{R.F.}}, \rho]$ yields:
$$
[H_{\text{R.F.}}, \rho] = \begin{pmatrix} \frac{f(t)}{2} (\rho_{10} - \rho_{01}) & (\Delta + \zeta(t)) \rho_{01} - \frac{f(t)}{2} (\rho_{11} - \rho_{00}) \\ -(\Delta + \zeta(t)) \rho_{10} + \frac{f(t)}{2} (\rho_{11} - \rho_{00}) & \frac{f(t)}{2} (\rho_{01} - \rho_{10}) \end{pmatrix}.
$$

Thus, the time derivative of $\rho$ is:
$$
\frac{d\rho}{dt} = -i \begin{pmatrix} \frac{f(t)}{2} (\rho_{10} - \rho_{01}) & (\Delta + \zeta(t)) \rho_{01} - \frac{f(t)}{2} (\rho_{11} - \rho_{00}) \\ -(\Delta + \zeta(t)) \rho_{10} + \frac{f(t)}{2} (\rho_{11} - \rho_{00}) & \frac{f(t)}{2} (\rho_{01} - \rho_{10}) \end{pmatrix}.
$$

This is the explicit form of $\frac{d\rho}{dt}$, derived by plugging in the commutator.

---

This completes the explicit derivation of the time evolution of the density matrix in the rotating frame.


### Section 1: Defining the Density Matrix Evolution Function

- `ρ00Re`, `ρ00Im`: Ground state population.
- `ρ01Re`, `ρ01Im`: Coherence between ground and excited states.
- `ρ10Re`, `ρ10Im`: Coherence conjugate, with Hermiticity ensuring $ρ10Re = ρ01Re$ and $ρ10Im = -ρ01Im$.

The input parameter vector `p` contains two critical parameters:
- `tg`: The gate time, which represents the duration of the quantum operation.

The function calculates the drive amplitude `ft`, which modulates the system dynamics:

This smooth modulation ensures that the amplitude evolves continuously over the gate time `tg`, avoiding abrupt changes.

#### 4. Defining the Matrices
The evolution of the density matrix is driven by real and imaginary contributions, captured in two matrices: `matRe` and `matIm`. These matrices are derived from the system's Hamiltonian and describe how each component of the density matrix changes over time.

- **`matRe`**: Contains terms representing the real contributions to the time evolution of the density matrix:
$$
\text{matRe} = \begin{bmatrix}
0.5(-ρ_{10Im} + ρ_{01Im}) f(t) & 0.5(ρ_{00Im} - ρ_{11Im}) f(t) \\
0.5(ρ_{11Im} - ρ_{00Im}) f(t) & 0.5(ρ_{10Im} - ρ_{01Im}) f(t)
\end{bmatrix}.
$$

- **`matIm`**: Contains terms representing the imaginary contributions to the time evolution:
$$
\text{matIm} = \begin{bmatrix}
0.5(ρ_{10Re} - ρ_{01Re}) f(t) & 0.5(ρ_{11Re} - ρ_{00Re}) f(t) \\
0.5(ρ_{00Re} - ρ_{11Re}) f(t) & 0.5(ρ_{01Re} - ρ_{10Re}) f(t)
\end{bmatrix}.
$$

#### 5. Assigning Derivatives
The derivatives of the density matrix components are computed from the entries of `matRe` and `matIm` and stored in the vector `du`. This vector represents the time derivatives of each component of the density matrix:
- `du[1] = matRe[1][1]`: Time derivative of $ρ_{00Re}$.
- `du[2] = matIm[1][1]`: Time derivative of $ρ_{00Im}$.
- `du[3] = matRe[1][2]`: Time derivative of $ρ_{01Re}$.
- `du[4] = matIm[1][2]`: Time derivative of $ρ_{01Im}$.
- `du[5] = matRe[2][1]`: Time derivative of $ρ_{10Re}$.
- `du[6] = matIm[2][1]`: Time derivative of $ρ_{10Im}$.
- `du[7] = matRe[2][2]`: Time derivative of $ρ_{11Re}$.
- `du[8] = matIm[2][2]`: Time derivative of $ρ_{11Im}$.

### Summary of the Function

This function `f` provides a complete model for the time evolution of the density matrix in a two-level quantum system. The function is foundational for simulating quantum operations and understanding system dynamics over time.


````julia
function f(du, u, p, t)
    
    # Unpack the components of the density matrix
    ρ00Re, ρ00Im, ρ01Re, ρ01Im, ρ10Re, ρ10Im, ρ11Re, ρ11Im = u
        
    # Unpack parameters
    tg, θ0 = p
    
    # Calculate the time-dependent drive amplitude
    ft = (θ0/tg)*(1-cos((2*pi*t)/tg))
    
    # Define the real matrix with previously provided simplifications
    matRe = [
            [0.5*(-ρ10Im + ρ01Im)*ft, 0.5*(ρ00Im - ρ11Im)*ft],
            [0.5*(ρ11Im - ρ00Im)*ft, 0.5*(ρ10Im - ρ01Im)*ft]
            ]
    
    # Define the imaginary matrix based on the derived terms
    matIm = [
            [0.5*(ρ10Re - ρ01Re)*ft, 0.5*(ρ11Re - ρ00Re)*ft],
            [0.5*(ρ00Re - ρ11Re)*ft, 0.5*(ρ01Re - ρ10Re)*ft]
            ]
    
    # Assign derivatives based on matRe and matIm entries
    du[1] = matRe[1][1]  # ̇ρ00Re
    du[2] = matIm[1][1]  # ̇ρ00Im
    du[3] = matRe[1][2]  # ̇ρ01Re
    du[4] = matIm[1][2]  # ̇ρ01Im
    du[5] = matRe[2][1]  # ̇ρ10Re
    du[6] = matIm[2][1]  # ̇ρ10Im
    du[7] = matRe[2][2]  # ̇ρ11Re
    du[8] = matIm[2][2]  # ̇ρ11Im
    
end
````

````julia
# Define the OUFixedEnsemble_2 function
function OUFixedEnsemble_2(τ_c, μ, σ, u0, tspan, dt0, solver, samp, reltol, abstol)

    function f(u, p, t)
        (1/τ_c)*(μ - u)
    end

    function g(u, p, t)
        sqrt(2*σ*σ/τ_c)
    end

    function prob_func(prob,i,repeat)
        remake(prob, u0 = rand(Normal(0,σ),1)[1])
    end
        
    prob = SDEProblem(f, g, u0, tspan)
    ensembleprob = EnsembleProblem(prob,prob_func=prob_func)
    sim = solve(ensembleprob, solver, EnsembleThreads(), trajectories=samp, saveat=dt0, abstol=abstol, reltol=reltol)
    return sim
end
````

This code computes the **autocorrelation function** for a set of stochastic trajectories. The autocorrelation function describes how the values of a stochastic process at one time are related to its values at a later time, providing insight into the temporal dependencies and memory of the system.

Here’s a breakdown of the code and its mathematical basis:

---

### **1. Inputs**

- `trajectories_matrix`: A matrix where each column corresponds to a trajectory of the stochastic process and each row is a time point.
- `mean_trajectory`: The mean of the trajectories at each time point (a vector of length equal to the number of rows in `trajectories_matrix`).
- `max_lag`: The maximum time lag ($ \tau $) for which the autocorrelation function will be computed.

---

### **2. Outputs**

- `covariances`: A vector containing the normalized autocorrelation values for lags from $ \tau = 0 $ to $ \tau = \text{max\_lag} $.

---

### **3. The Mathematical Basis**

The autocorrelation at lag $ \tau $ is defined as:

$$
R(\tau) = \frac{\text{Cov}(X(t), X(t+\tau))}{\text{Var}(X(t))}
$$

Where:
- $ \text{Cov}(X(t), X(t+\tau)) $ is the covariance of the process at times $ t $ and $ t+\tau $.
- $ \text{Var}(X(t)) $ is the variance of the process at time $ t $.

The numerator, $ \text{Cov}(X(t), X(t+\tau)) $, measures the linear dependency between $ X(t) $ and $ X(t+\tau) $. The function normalizes by $ \text{Var}(X(t)) $ (the value at $ \tau = 0 $) to give a dimensionless correlation measure between $ -1 $ and $ 1 $.

#### Key Steps in the Code:
1. **Center the Data**:
   $$
   \text{diff\_matrix} = \text{trajectories\_matrix} - \text{mean\_trajectory}
   $$
   This centers each trajectory around its mean at each time point.

2. **Iterate Over Time Lags ($ \tau $)**:
   For each lag $ \tau $ (from 0 to `max_lag`):
   - Compute the covariance between the trajectory values at time $ t $ and at time $ t + \tau $ by:
     $$
     \text{cov}_{\tau} = \frac{1}{N} \sum_{t=1}^{n-\tau} \text{mean}( \text{diff\_matrix}[t, :] \cdot \text{diff\_matrix}[t+\tau, :] )
     $$
     - $ \text{diff\_matrix}[t, :] $: The deviation of all trajectories at time $ t $ from their mean.
     - $ \text{diff\_matrix}[t+\tau, :] $: The deviation of all trajectories at time $ t + \tau $ from their mean.
     - The dot product multiplies these deviations pointwise for each trajectory, computing a cross-term.

3. **Normalize**:
   The autocorrelation values are normalized by $ \text{cov}_0 $ (the variance at lag $ 0 $) to ensure $ R(0) = 1 $.

---

### **4. Key Computational Steps in the Code**

1. **Initialize the Covariance Vector**:
   The `covariances` vector stores the autocorrelation values for $ \tau = 0, 1, \dots, \text{max\_lag} $.

2. **Double Loop Over Time Lags and Time Points**:
   - Outer loop: Iterates over the time lags $ \tau $.
   - Inner loop: Iterates over the time points $ t $, ensuring $ t+\tau $ does not exceed the trajectory length.

3. **Aggregate Cross-Terms**:
   For each $ \tau $, compute the covariance by summing the product of deviations over all trajectories and normalizing by the number of valid time points ($ n - \tau $).

4. **Normalize to Create Autocorrelation**:
   Normalize all computed covariances by the value at lag $ \tau = 0 $ (the variance), yielding the autocorrelation function.

---

### **5. Why is Normalization Important?**

The normalization ensures that the autocorrelation values fall between $ -1 $ and $ 1 $ and that $ R(0) = 1 $, making it easier to interpret the strength and direction of the temporal dependencies:

- $ R(\tau) \approx 1 $: Strong positive correlation at lag $ \tau $.
- $ R(\tau) \approx -1 $: Strong negative correlation at lag $ \tau $.
- $ R(\tau) \approx 0 $: No correlation at lag $ \tau $.

---

### **6. Example Use Case**

This function is typically used in time-series analysis to understand the memory and time-dependence of stochastic processes, such as:

- Identifying time scales of decay for autocorrelations in stochastic dynamics.
- Characterizing processes like Ornstein-Uhlenbeck dynamics or Brownian motion.

---

### **Summary of How Noise Is Handled**
The autocorrelation reflects how the stochastic noise affects the process's memory. By analyzing how quickly the autocorrelation decays, one can infer the system's temporal dynamics and how long it "remembers" past fluctuations.


````julia
# Compute the covariance function
function autocorrelation_function(trajectories_matrix, mean_trajectory, max_lag)
    n = size(trajectories_matrix, 1)  # number of time points
    covariances = Vector{Float64}(undef, max_lag+1)
    diff_matrix = (trajectories_matrix.- mean_trajectory)
    
    for τ in 0:max_lag
        cov_τ = 0.0
        for t in 1:(n - τ)
            cov_τ += mean(diff_matrix[t, :].*diff_matrix[t+τ,:])
        end
        covariances[τ + 1] = cov_τ / (n - τ)
    end
    return covariances/covariances[1]
end
````

````julia
function noisy_qubit(p, u, tspan, dt, solver, samp, reltol, abstol) 
    
    function f(du, u, p, t)
        ρ00Re, ρ00Im, ρ01Re, ρ01Im, ρ10Re, ρ10Im, ρ11Re, ρ11Im, ξ, ζ = u

        θx, θy, μx, μy, σx, σy, tg, θ0, Δ = p

        
        
        # Define the derivatives for the real and imaginary parts
        ft = (θ0/tg)*(1 - cos((2*pi*t)/tg))

        # Define the real matrix with previously provided simplifications
        matRe = [
                [0.5*(-ρ10Im + ρ01Im)*ft, -(Δ+ξ)*ρ01Im + 0.5*(ρ00Im - ρ11Im)*ft],
                [(Δ+ξ)*ρ10Im + 0.5*(ρ11Im - ρ00Im)*ft, 0.5*(ρ10Im - ρ01Im)*ft]
                ]

        matIm = [
                [0.5*(ρ10Re - ρ01Re)*ft, (Δ+ξ)*ρ01Re + 0.5*(ρ11Re - ρ00Re)*ft],
                [-(Δ+ξ)*ρ10Re + 0.5*(ρ00Re - ρ11Re)*ft, 0.5*(ρ01Re - ρ10Re)*ft]
                ]
        
        du[1] = matRe[1][1]  # ̇ρ00Re
        du[2] = matIm[1][1]  # ̇ρ00Im
        du[3] = matRe[1][2]  # ̇ρ01Re
        du[4] = matIm[1][2]  # ̇ρ01Im
        du[5] = matRe[2][1]  # ̇ρ10Re
        du[6] = matIm[2][1]  # ̇ρ10Im
        du[7] = matRe[2][2]  # ̇ρ11Re
        du[8] = matIm[2][2]  # ̇ρ11Im
        du[9] = θx*(μx-ξ) # noise produced by the OU Process
        #du[10] = θy*(μy-ζ)+σy*ξ # noise produced by coupuling to the OU Proccess

    end

    function g(du, u, p, t)

        θx, θy, μx, μy, σx, σy, tg, θ0 = p
        
        du[9] = σx
        #du[10] = 0
    end

    prob = SDEProblem(f, g, u0, tspan, p)
    ensembleprob = EnsembleProblem(prob)
    sim = solve(ensembleprob, dt=dt, solver, EnsembleThreads(), trajectories=samp, abstol=abstol, reltol=reltol)
    
    return sim
end            
````

````julia
function f(du, u, p, t)
    L11Re, L11Im, L12Re, L12Im, L13Re, L13Im, L14Re, L14Im, L21Re, L21Im, L22Re, L22Im, L23Re, L23Im, L24Re, L24Im, L31Re, L31Im, L32Re, L32Im, L33Re, L33Im, L34Re, L34Im, L41Re, L41Im, L42Re, L42Im, L43Re, L43Im, L44Re, L44Im, ξ = u

    θx, μx, σx, tg, θ0, Δ = p

    # Define the derivatives for the real and imaginary parts
    ft = (θ0/tg)*(1-cos((2*pi*t)/tg))

    # Define the real matrix with previously provided simplifications
    matRe = [
                [
                    0.5 * (L21Im - L31Im) * ft, 
                    0.5 * (L22Im - L32Im) * ft, 
                    0.5 * (L23Im - L33Im) * ft, 
                    0.5 * (L24Im - L34Im) * ft
                ],
                [
                    -L21Im * (Δ + ξ) + 0.5 * (L11Im - L41Im) * ft, 
                    -L22Im * (Δ + ξ) + 0.5 * (L12Im - L42Im) * ft, 
                    -L23Im * (Δ + ξ) + 0.5 * (L13Im - L43Im) * ft, 
                    -L24Im * (Δ + ξ) + 0.5 * (L14Im - L44Im) * ft
                ],
                [
                    L31Im * (Δ + ξ) - 0.5 * (L11Im - L41Im) * ft, 
                    L32Im * (Δ + ξ) - 0.5 * (L12Im - L42Im) * ft, 
                    L33Im * (Δ + ξ) - 0.5 * (L13Im - L43Im) * ft, 
                    L34Im * (Δ + ξ) - 0.5 * (L14Im - L44Im) * ft
                ],
                [
                    0.5 * (-L21Im + L31Im) * ft, 
                    0.5 * (-L22Im + L32Im) * ft, 
                    0.5 * (-L23Im + L33Im) * ft, 
                    0.5 * (-L24Im + L34Im) * ft
                ]
            ] 

    matIm = [
                [
                    0.5 * (-L21Re + L31Re) * ft, 
                    0.5 * (-L22Re + L32Re) * ft, 
                    0.5 * (-L23Re + L33Re) * ft, 
                    0.5 * (-L24Re + L34Re) * ft
                ],
                [
                    L21Re * (Δ + ξ) - 0.5 * (L11Re - L41Re) * ft, 
                    L22Re * (Δ + ξ) - 0.5 * (L12Re - L42Re) * ft, 
                    L23Re * (Δ + ξ) - 0.5 * (L13Re - L43Re) * ft, 
                    L24Re * (Δ + ξ) - 0.5 * (L14Re - L44Re) * ft
                ],
                [
                    -L31Re * (Δ + ξ) + 0.5 * (L11Re - L41Re) * ft, 
                    -L32Re * (Δ + ξ) + 0.5 * (L12Re - L42Re) * ft, 
                    -L33Re * (Δ + ξ) + 0.5 * (L13Re - L43Re) * ft, 
                    -L34Re * (Δ + ξ) + 0.5 * (L14Re - L44Re) * ft
                ],
                [
                    0.5 * (L21Re - L31Re) * ft, 
                    0.5 * (L22Re - L32Re) * ft, 
                    0.5 * (L23Re - L33Re) * ft, 
                    0.5 * (L24Re - L34Re) * ft
                ]
            ]
    
    du[1]  = matRe[1][1]  # ̇L11Re
    du[2]  = matIm[1][1]  # ̇L11Im
    du[3]  = matRe[1][2]  # ̇L12Re
    du[4]  = matIm[1][2]  # ̇L12Im
    du[5]  = matRe[1][3]  # ̇L13Re
    du[6]  = matIm[1][3]  # ̇L13Im
    du[7]  = matRe[1][4]  # ̇L14Re
    du[8]  = matIm[1][4]  # ̇L14Im
    du[9]  = matRe[2][1]  # ̇L21Re
    du[10] = matIm[2][1]  # ̇L21Im
    du[11] = matRe[2][2]  # ̇L22Re
    du[12] = matIm[2][2]  # ̇L22Im
    du[13] = matRe[2][3]  # ̇L23Re
    du[14] = matIm[2][3]  # ̇L23Im
    du[15] = matRe[2][4]  # ̇L24Re
    du[16] = matIm[2][4]  # ̇L24Im
    du[17] = matRe[3][1]  # ̇L31Re
    du[18] = matIm[3][1]  # ̇L31Im
    du[19] = matRe[3][2]  # ̇L32Re
    du[20] = matIm[3][2]  # ̇L32Im
    du[21] = matRe[3][3]  # ̇L33Re
    du[22] = matIm[3][3]  # ̇L33Im
    du[23] = matRe[3][4]  # ̇L34Re
    du[24] = matIm[3][4]  # ̇L34Im
    du[25] = matRe[4][1]  # ̇L41Re
    du[26] = matIm[4][1]  # ̇L41Im
    du[27] = matRe[4][2]  # ̇L42Re
    du[28] = matIm[4][2]  # ̇L42Im
    du[29] = matRe[4][3]  # ̇L43Re
    du[30] = matIm[4][3]  # ̇L43Im
    du[31] = matRe[4][4]  # ̇L44Re
    du[32] = matIm[4][4]  # ̇L44Im
    du[33] = θx*(μx-ξ) # noise produced by the OU Process
end

function g(du, u, p, t)

        θx, μx, σx, tg, θ0, Δ = p
        
        du[33] = σx
end

# Set the initial conditions
L11Re0 = 1.0  # Initial real part of L11 (identity matrix assumption)
L11Im0 = 0.0  # Initial imaginary part of L11
L12Re0 = 0.0  # Initial real part of L12
L12Im0 = 0.0  # Initial imaginary part of L12
L13Re0 = 0.0  # Initial real part of L13
L13Im0 = 0.0  # Initial imaginary part of L13
L14Re0 = 0.0  # Initial real part of L14
L14Im0 = 0.0  # Initial imaginary part of L14
L21Re0 = 0.0  # Initial real part of L21
L21Im0 = 0.0  # Initial imaginary part of L21
L22Re0 = 1.0  # Initial real part of L22 (identity matrix assumption)
L22Im0 = 0.0  # Initial imaginary part of L22
L23Re0 = 0.0  # Initial real part of L23
L23Im0 = 0.0  # Initial imaginary part of L23
L24Re0 = 0.0  # Initial real part of L24
L24Im0 = 0.0  # Initial imaginary part of L24
L31Re0 = 0.0  # Initial real part of L31
L31Im0 = 0.0  # Initial imaginary part of L31
L32Re0 = 0.0  # Initial real part of L32
L32Im0 = 0.0  # Initial imaginary part of L32
L33Re0 = 1.0  # Initial real part of L33 (identity matrix assumption)
L33Im0 = 0.0  # Initial imaginary part of L33
L34Re0 = 0.0  # Initial real part of L34
L34Im0 = 0.0  # Initial imaginary part of L34
L41Re0 = 0.0  # Initial real part of L41
L41Im0 = 0.0  # Initial imaginary part of L41
L42Re0 = 0.0  # Initial real part of L42
L42Im0 = 0.0  # Initial imaginary part of L42
L43Re0 = 0.0  # Initial real part of L43
L43Im0 = 0.0  # Initial imaginary part of L43
L44Re0 = 1.0  # Initial real part of L44 (identity matrix assumption)
L44Im0 = 0.0  # Initial imaginary part of L44

u1 = [L11Re0, L11Im0, L12Re0, L12Im0, L13Re0, L13Im0, L14Re0, L14Im0,L21Re0, L21Im0, L22Re0, L22Im0, L23Re0, L23Im0, L24Re0, L24Im0,L31Re0, L31Im0, L32Re0, L32Im0, L33Re0, L33Im0, L34Re0, L34Im0,L41Re0, L41Im0, L42Re0, L42Im0, L43Re0, L43Im0, L44Re0, L44Im0, 0.0]
# Parameters deffining the noise and the hamiltonion
τ_c = 1
σ = sqrt(1/(10000*τ_c))
θx = -1/τ_c
μx = 0.0
σx = sqrt(2*σ^2/τ_c)
tg = τ_c/10
θ0 = pi/2
Δ = 0

p2 = [θx, μx, σx, tg, θ0, Δ]
tspan = (0.0, tg) # Set the time span to solve the ODEs over (t0, tg)

# Set the time interval and define the conditions of the solver
saveint = tg/100
solver = EM()
samples = 10000
relto = 1.0 * 10^(-10)
absol = 1.0 * 10^(-10)

prob = SDEProblem(f, g, u1, tspan, p2)
ensembleprob = EnsembleProblem(prob)
sim = solve(ensembleprob, LambaEulerHeun(), EnsembleThreads(), trajectories=samples, abstol=absol, reltol=relto,save_everystep = false, save_start = false)
````

````julia
function average_fidel_per_sigma(L_end_per_sigma)
    L = L_end_per_sigma
    average_fidelity_per_sig = []
    
    for i in 1:size(L,1)
        LRe_mean_noise = [L[i][1] L[i][3] L[i][5] L[i][7];
                         L[i][9] L[i][11] L[i][13] L[i][15];
                         L[i][17] L[i][19] L[i][21] L[i][23];
                         L[i][25] L[i][27] L[i][29] L[i][31]]
        LIm_mean_noise = [L[i][2]  L[i][4]  L[i][6]  L[i][8];
                         L[i][10] L[i][12] L[i][14] L[i][16];
                         L[i][18] L[i][20] L[i][22] L[i][24];
                         L[i][26] L[i][28] L[i][30] L[i][32]]
    
        L_mean = LRe_mean_noise + LIm_mean_noise*im
    
        rho_ideal_list = [UI*ρ_zp*UI', UI*ρ_zm*UI', UI*ρ_xp*UI', UI*ρ_xm*UI', UI*ρ_yp*UI', UI*ρ_ym*UI']
        rho_noisy_list = [reshape(L_mean*vec(ρ_zp),2,2), reshape(L_mean*vec(ρ_zm),2,2), reshape(L_mean*vec(ρ_xp),2,2), reshape(L_mean*vec(ρ_xm),2,2), reshape(L_mean*vec(ρ_yp),2,2), reshape(L_mean*vec(ρ_ym),2,2)]
    
        append!(average_fidelity_per_sig, average_fidelity(rho_ideal_list, rho_noisy_list))
    end
    return average_fidelity_per_sig
    
end
    
````
